```python
from alloskin.common.utils import load_config
from alloskin.pipeline.runner import run_analysis

config_file = "config/residue_selection.yaml"
active_traj_file = "/storage_common/angiod/A2A/vdamore/FAN.5000.xtc"
active_topo_file = "/storage_common/angiod/A2A/vdamore/FAN.pdb"
inactive_traj_file = "/storage_common/angiod/A2A/vdamore/INapo.5000.xtc"
inactive_topo_file = "/storage_common/angiod/A2A/vdamore/INapo.pdb"
active_slice = ":1000"
inactive_slice = ":1000"
num_workers = 64
analysis_type = "qubo"
filter_top_n = 80

static_params = {
    "static_method": 'rf'
}

qubo_params = {
    "qubo_method": 'ii',
    "active_topo_file": active_topo_file,
    "static_results_path": "/home/angiod@usi.ch/AllosKin/data/results/03c3c53a-643f-4150-9bcc-e616df82d96e.json",
    "num_solutions": 5,
    "qubo_cv_folds": 3,
    "qubo_n_estimators": 50,
    "alpha_size": 0.8, # Penalty for set size. Ti seleziono se predici almeno un altro residuo con ii < 0.2
    "beta_coverage": 1.0, # Reward for predicting others
    "lambda_redundancy": 2.0, # Penalty for symmetric redundancy
    "filter_top_n": 20,
}

# ---

config = load_config(config_file=config_file)
residue_selections = config.get('residue_selections')

file_paths = {
    'active_traj': active_traj_file,
    'active_topo':active_topo_file,
    'inactive_traj': inactive_traj_file,
    'inactive_topo': inactive_topo_file,
}

params = {
    "active_slice": active_slice,
    "inactive_slice": inactive_slice,
    "num_workers": num_workers,
}
params.update(static_params)
params.update(qubo_params)

results, mapping = run_analysis(
    analysis_type=analysis_type,
    file_paths=file_paths,
    params=params,
    residue_selections=residue_selections
)
```
